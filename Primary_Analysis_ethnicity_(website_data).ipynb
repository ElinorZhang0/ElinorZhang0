# San Mateo | Santa Clara county ethnic stats (old)

## Data Clean

```python
# read table 
counties_summary=pd.read_csv('PL94_2020_Adjusted_P24_DOJ_counties_summary.csv')
pd.options.display.max_columns = None
pd.options.display.max_rows = None
print(counties_summary.shape)
counties_summary.head()
```

```python
counties_summary.columns
```

```python
col = ['COUNTY20','NAME','Population','Hispanic Origin','Not Hispanic','NH_1 Race', 'NH_Wht', 'NH_Blk', 'NH_Ind', 'NH_Asn', 'NH_Hwn','NH_Oth']
counties_summary_one_race = pd.DataFrame(counties_summary,columns=col)
print(counties_summary_one_race.shape,type(counties_summary_one_race))
counties_summary_one_race.head()
```

```python
col1 = ['NAME','Hispanic Origin','NH_Wht', 'NH_Blk', 'NH_Ind', 'NH_Asn', 'NH_Hwn','NH_Oth']
counties_summary_one_race_stats = pd.DataFrame(counties_summary,columns=col1)
print(counties_summary_one_race_stats.shape)
counties_summary_one_race_stats.head()
```

## Santa Clara

```python
Santa_Clara_County = counties_summary_one_race_stats.loc[counties_summary_one_race['NAME']=='Santa Clara County']
Santa_Clara_County
```

```python
Santa_Clara_County = Santa_Clara_County.set_index('NAME')
Santa_Clara_County_percentage = Santa_Clara_County.div(Santa_Clara_County.sum(axis=1)/100, axis=0) 
Santa_Clara_County_percentage = Santa_Clara_County_percentage.rename(columns={'NH_Blk': '%Black CVAP','NH_Wht': '%White CVAP','NH_Asn': '%Asian CVAP'})
Santa_Clara_County_percentage
```
